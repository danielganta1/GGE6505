```python
actors_url = "/content/actors.csv"
actors_data = pd.read_csv(actors_url)
display(actors_data)
actors_data.isnull().sum()
```

```python
actors_data.drop(actors_data.iloc[:, 9:25], inplace=True, axis=1)
display(actors_data)
actors_data.isnull().sum()
```

```python
import numpy as np
actors_data['gen'] = actors_data['gen'].fillna('NA')
conditions = [
    actors_data['gen'] == 'M',
    actors_data['gen'] == 'F',
    actors_data['gen'] == 'NA'
]

choice = ['M','F','NA']

actors_data['gen'] = np.select(conditions,choice)
actors_data.isnull().sum()
```

```python
actors_data
actors_data.to_csv('/content/actors.csv', index=False)
```

```python
data_column_category = actors_data.select_dtypes(exclude=[np.number]).columns
# data_column_category
actors_data[data_column_category].head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

for i in data_column_category:
    actors_data[i] = label_encoder.fit_transform(actors_data[i])

actors_data.head()
```

```python
actors_data
```

```python
from sklearn import preprocessing

scaler = preprocessing.MinMaxScaler()
names = actors_data.columns
d = scaler.fit_transform(actors_data)
scaled_df = pd.DataFrame(d, columns=names)
scaled_df.head()
```

```python
import scipy.stats as stats
scaled_df['dob_zscore'] = stats.zscore(scaled_df['dob'])
display(scaled_df)
```
